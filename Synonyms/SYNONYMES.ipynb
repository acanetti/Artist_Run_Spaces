```python
import nltk
nltk.download('wordnet')
def synonym(txt):  
    synonyms=[]
    from nltk.corpus import wordnet
    for syn in wordnet.synsets(txt):
        for l in syn.lemmas():
            synonyms.append(l.name())
    return set(synonyms)
```

```python
synonym('slut')
```

```python
Unique=set([])
for i in data_trad['clean_presentation_trad'].values: 
    for j in i : 
        Unique.add(j)
```

```python
Dict={key : synonym(key) for key in Unique}
```
