```python
def tr(x):
    translator = googletrans.Translator()  # Instance du traducteur
    try : 
        return  translator.translate(x,dest='en',src='fr').text  
    except : 
        pass 
        #fr -> en 
def translate_all(data,col='presentation'): 
    #googletrans recquiered : 
    dis=data[data.lang=='fr']['presentation']
    N=len(dis)
    to_translate=[] 
    index=dis.index
    itera = 1
    for i in dis.values:  
        to_translate.append(tr(i)) 
        print (f' set n°{itera} on {N}')
        itera +=1
        
    return to_translate , index   
        
        
        
    
    
```
